```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from time import sleep
from bs4 import BeautifulSoup
import re
import sys
import requests
import json
import datetime
import pandas as pd
import pickle
```

```python
  
def store(news_timetext):
    with open("ltn_timetext.pickle", "wb") as f:
        pickle.dump(news_timetext, f)

def ltn(search, news_timetext):
    options = webdriver.ChromeOptions()
    options.add_argument('--ignore-certificate-errors')
    # options.add_argument('--proxy-server=101.9.200.184')
    options.add_argument('--incognito')
    options.add_argument('headless')                        
    driver = webdriver.Chrome(chrome_options=options)
    driver.set_page_load_timeout(60)
    url = f"https://news.ltn.com.tw/topic/{search}"
    print(url)
    driver.get(url)
    driver.find_element(by=By.CLASS_NAME, value="softPush_refuse").click()
    print("==================================scroll==================================")
    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        times_t = driver.find_elements(by=By.CLASS_NAME, value="time")
        times = [pd.to_datetime(i.text) for i in times_t]
        print(min(times))
        sleep(3)
        if min(times) <= pd.to_datetime("2023/01/01"):
            break
        # print(min(driver.find_elements(by=By.CLASS_NAME, value="time")))
    news_ps = driver.find_elements(by=By.CLASS_NAME, value="tit")  
    for news_p in news_ps:
        print("==================================new==================================")
        driver = webdriver.Chrome(chrome_options=options)
        driver.get(news_p.get_attribute("href"))
        time = driver.find_element(by=By.CLASS_NAME, value="time").text
        text = driver.find_element(by=By.CLASS_NAME, value="whitecon").text
        text = text[:text.find("一手掌握經濟脈動 點我訂閱自由財經Youtube頻道")].replace("\n", "").replace("\r", "")
        print(time)
        print(text)
        if time in news_timetext and text in news_timetext[time][0]:
            print("duplicate")
            continue
        news_timetext.update({time: text})
        sleep(5)
        store(news_timetext)
    return news_timetext
    # driver.find_element(by=By.ID, value="APjFqb").send_keys(Keys.ENTER)
    # driver.find_element(by=By.CLASS_NAME, value="zItAnd").click() # search with href?
```
